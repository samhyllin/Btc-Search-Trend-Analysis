```python
df_btc_price['DATE'] = pd.to_datetime(df_btc_price['DATE'], errors='coerce')
df_btc_price.dtypes
```

convertion of DATE column from object to datetime

```python
max_search = df_btc_search['BTC_NEWS_SEARCH'].max()
min_search = df_btc_search['BTC_NEWS_SEARCH'].min()

print(f"The largest Bitcoin news search value is {max_search} and the smallest is {min_search}")
```

```python
df_btc_monthly = df_btc_price.resample('ME', on='DATE').last()
df_btc_monthly.head()
```

This will give us the available price of the month - the price at month-end.

```python
df_btc_price['YEAR'] = df_btc_price['DATE'].dt.year
df_btc_yearly = df_btc_price.groupby('YEAR')[['CLOSE','VOLUME']].mean().reset_index()
df_btc_yearly.head()
```

This converts daily Bitcoin data into yearly averages.

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_btc_yearly['YEAR'],
    y=df_btc_yearly['CLOSE'],
    name='Price',
    line=dict(color='orange'),
    yaxis='y1',
    hovertemplate='Year: %{x}<br>Price: %{y}<extra></extra>'
))

fig.add_trace(go.Bar(
    x=df_btc_yearly['YEAR'],
    y=df_btc_yearly['VOLUME'],
    name='Volume',
    marker_color='skyblue',
    yaxis='y2',
    hovertemplate='Year: %{x}<br>Volume: %{y}<extra></extra>'
))

fig.update_layout(
    title='Yearly Bitcoin Price and Volume',
    xaxis_title='Year',
    yaxis=dict(
        title='Price',
        side='left'
    ),
    yaxis2=dict(
        title='Volume',
        overlaying='y',
        side='right'
    ),
    legend=dict(x=0, y=1),
    template='plotly_white'
)

fig.show()
```

This chart shows Bitcoin's yearly average closing price (line) and average trading volume (bars). Dual axes highlight trends in price and trading activity. Aggregating by year smooths daily fluctuations, making long-term patterns easy to spot. Hover over each point for exact values.

```python
df_btc_search['MONTH'] = pd.to_datetime(df_btc_search['MONTH'])


df_btc_search['year'] = df_btc_search['MONTH'].dt.year
yearly_btc_search = df_btc_search.groupby('year')['BTC_NEWS_SEARCH'].sum()


print(yearly_btc_search)
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Bar(
            x=yearly_btc_search.index,
            y=yearly_btc_search.values,
            marker_color='#3A7BD5',  
            text=yearly_btc_search.values,
            textposition='auto',
            hovertemplate=
                '<b>Year:</b> %{x}<br>' +
                '<b>Total Searches:</b> %{y}<extra></extra>',
        )
    ]
)

fig.update_layout(
    title='Total Bitcoin Search Volume per Year',
    xaxis_title='Year',
    yaxis_title='Total BTC Searches',
    title_font_size=22,
    xaxis_title_font_size=16,
    yaxis_title_font_size=16,
    xaxis_tickfont_size=14,
    yaxis_tickfont_size=14,
    template='plotly_white',
    width=900,
    height=500,
    bargap=0.2
)

fig.show()

```

This chart sums Bitcoin news searches by year, showing how public interest has changed over time. The gradient blue bars and interactive hover make it easy to spot peak years and trends at a glance.

```python
import matplotlib.dates as mdates

plt.figure(figsize=(14,8), dpi=120)
plt.title('BTC Web Search vs Price', fontsize=18)

ax1 = plt.gca()
ax2 = ax1.twinx()

years = mdates.YearLocator()
years_fmt = mdates.DateFormatter('%Y')
months = mdates.MonthLocator()

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1.set_ylabel('BTC Price', color='#FF5733', fontsize=14)
ax2.set_ylabel('Search Trend', color='#4A90E2', fontsize=14)

ax1.plot(df_btc_monthly.index, df_btc_monthly.CLOSE,
         color='#FF5733', linewidth=3)
ax2.plot(df_btc_monthly.index, df_btc_search.BTC_NEWS_SEARCH,
         color='#4A90E2', linewidth=3)

plt.show()
```

What the Plot Shows

The chart compares Bitcoins monthly price with its monthly news search volume. Both are plotted on separate y-axes so you can see how they move relative to each other over time

A few things jump out:

When Bitcoin becomes a hot topic, you see search interest rise at the same time its price rises.

The massive spike in late 2017 and early 2018 shows this clearly. The price shot up fast and search popularity exploded because everyone was curious, confused, or trying to get involved.

In early 2019, the price jumped again, but the search volume did not rise as much. People already knew Bitcoin by then, so fewer people were searching even though the price was moving.

Overall, the plot suggests that search interest sometimes rises with price movements, especially during hype periods. It also shows that as Bitcoin becomes more familiar, big price changes do not always create big spikes in searches.
